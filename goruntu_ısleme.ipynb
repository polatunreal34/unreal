```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

bgr = cv2.imread("/content/5956322790430853401.jpg")
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

h, w, c = rgb.shape
print(h, w, c)

plt.imshow(rgb)
plt.axis("off")  # eksenleri kapat


```

```python
gri_para = cv2.cvtColor(bulanÄ±k_para, cv2.COLOR_RGB2GRAY)
plt.imshow(gri_para, cmap="gray")
plt.axis("off")
```

```python
bulanÄ±k_para = cv2.GaussianBlur(rgb, (7,7), 0)
plt.imshow(bulanÄ±k_para)
plt.axis("off")
```

```python
_, binary_para = cv2.threshold(gri_para, 200, 200, cv2.THRESH_BINARY)
plt.imshow(binary_para, cmap="gray")
plt.axis("off")
```

```python
kenar_para = cv2.Canny(binary_para, 127, 255)
plt.imshow(kenar_para, cmap="gray")
plt.axis("off")
```

```python
contours, hierarchy = cv2.findContours(kenar_para,  cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
kontorlu_para = cv2.drawContours(rgb, contours, -1, (255,0,0), 4)

for i, kontor in enumerate(contours,start=1):
  if cv2.contourArea(kontor) > 180:
    x, y, w, h = cv2.boundingRect(kontor)
    cv2.rectangle(kontorlu_para, (x,y), (x+w, y+h), (0,255,0), 4)
    cv2.putText(kontorlu_para, str(i), (x,y), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255),4)
    cv2.putText(kontorlu_para, "P D", (x+20,y+60), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255),4)
    print(cv2.contourArea(kontor))
plt.imshow(kontorlu_para)
plt.axis("off")
```
